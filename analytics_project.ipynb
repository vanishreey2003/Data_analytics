importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

importing the dataset

```python
dataset=pd.read_csv('https://raw.githubusercontent.com/arora123/Data/master/Automobile_data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

taking care of missing data

```python
dataset.isnull().sum()
```

yes there are missing values in price

drop with missing value

```python
dataset=dataset.dropna()
```

```python
print(dataset)
```

1.cars with minimum or maximum price

```python
min_price_car=dataset[dataset['price']==dataset['price'].min()]
max_price_car=dataset[dataset['price']==dataset['price'].max()]
print("car with minimum price:\n",min_price_car)
print("car with maximum price:\n",max_price_car)
```

2. cars with different body-styles

```python
body_style_count=dataset['body-style'].value_counts()
print("number of cars per body style:\n",body_style_count)
```

3. average price of all body types of car (descending oreder):

```python
average_price_body_type=dataset.groupby('body-style')['price'].mean().sort_values(ascending=False)
print("average price of all body types of car (decending):\n",average_price_body_type)
```

4. average price of cars by company (decending order):

```python
average_price_by_company=dataset.groupby('company')['price'].mean().sort_values(ascending=False)
print("average price of cars by different companies (decending):\n",average_price_by_company)
```

5. count car types having different number cylinders

```python
cylinder_counts=dataset['num-of-cylinders'].value_counts()
print("count of car types having different number cylinders:\n",cylinder_counts)
```

exploratory data analysis and visualization

1. company and price

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(dataset['company'],dataset['price'])
plt.xlabel('company')
plt.ylabel('price')
plt.title('company vs price')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

2.company and no of cylinders

```python
plt.figure(figsize=(10,6))
dataset.groupby('company')['num-of-cylinders'].value_counts().unstack().plot(kind='bar',stacked=True)
plt.xlabel('company')
plt.ylabel('count')
plt.title('company vs no of cylinders')
plt.xticks(rotation=90)
plt.legend(title='no of cylinders')
plt.tight_layout()
plt.show()
```

3. body type and price

```python
plt.figure(figsize=(10,6))
plt.scatter(dataset['body-style'],dataset['price'])
plt.xlabel('Body type')
plt.ylabel('Price')
plt.title('Body type vs Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

4. body type and cylinders

```python
plt.figure(figsize=(10,6))
dataset.groupby('body-style')['num-of-cylinders'].value_counts().unstack().plot(kind='bar',stacked=True)
plt.xlabel('Body type')
plt.ylabel('count')
plt.title('Body type vs number of Cylinders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. engine type and horsepower

```python
plt.figure(figsize=(10,6))
plt.scatter(dataset['engine-type'],dataset['horsepower'])
plt.xlabel('Engine type')
plt.ylabel('Horsepower')
plt.title('Engine type vs Horsepower')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
